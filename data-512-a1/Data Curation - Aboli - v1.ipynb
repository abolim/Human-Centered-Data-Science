**Goal:** The goal of this notebook is to construct, analyze, and publish a dataset of monthly traffic on English Wikipedia from January 1 2008 through August 31 2020. 

### Step 1: Gathering the data

In order to **measure Wikipedia traffic from 01-Jan-2008 to 31-Aug-2020**, we will collect data from two different API endpoints, the Legacy Pagecounts API and the Pageviews API.

**The Legacy Pagecounts API** provides access to desktop and mobile traffic data from December 2007 through July 2016. <br>
__[Documentation](https://wikitech.wikimedia.org/wiki/Analytics/AQS/Legacy_Pagecounts)__ <br>
__[API endpoint](https://wikimedia.org/api/rest_v1/#/Pagecounts_data_(legacy)/get_metrics_legacy_pagecounts_aggregate_project_access_site_granularity_start_end)__

**The Pageviews API** provides access to desktop, mobile web, and mobile app traffic data from July 2015 through Aug 2020.<br>
__[Documentation](https://wikitech.wikimedia.org/wiki/Analytics/AQS/Pageviews)__ <br>
__[API endpoint](https://wikimedia.org/api/rest_v1/#/Pageviews_data/get_metrics_pageviews_aggregate_project_access_agent_granularity_start_end)__

For each API, we will collect data for all months where data is available and then save the raw results into 5 separate JSON source data files (one file per API query type) for our analysis.

```python
# Import libraries

import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

We will create variables to store parameterized URLs to access the API endpoints

```python
# Create variables for parameterized API endpoint URLs

endpoint_legacy = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'

endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'
```

Now we will define the parameters and function for accessing the 5 different APIs for:
1. Legacy desktop page counts
2. Legacy mobile page counts
3. Current desktop page views
4. Current mobile app page views
5. Current mobile web page views

```python
# Create parameters for above URLs to access the API endpoints

# Monthly aggregated legacy page counts data from 01-Jan-2008 to 31-Jul-2016
# Refer: https://wikimedia.org/api/rest_v1/#!/Legacy_data/get_metrics_legacy_pagecounts_aggregate_project_access_site_granularity_start_end
# For end use 1st day of month following final month of data (here: 2016080100)

params_pagecounts_desktop_site = {"project" : "en.wikipedia.org",
                 "access-site" : "desktop-site",
                 "granularity" : "monthly",
                 "start" : "2008010100",                
                 "end" : "2016080100" 
                    }

params_pagecounts_mobile_site = {"project" : "en.wikipedia.org",
                 "access-site" : "mobile-site",
                 "granularity" : "monthly",
                 "start" : "2008010100",                
                 "end" : "2016080100"
                    }

# Monthly aggregated current standard pageview data from 01-July-2015 to 31-Aug-2020
# Refer: https://wikimedia.org/api/rest_v1/#!/Pageviews_data/get_metrics_pageviews_aggregate_project_access_agent_granularity_start_end
# Filtering for agent = user as we are interested in the organic traffic only

params_pageviews_desktop = {"project" : "en.wikipedia.org",
                    "access" : "desktop",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    "end" : '2020090100'
                        }

params_pageviews_mobile_app = {"project" : "en.wikipedia.org",
                    "access" : "mobile-app",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    "end" : '2020090100'
                        }

params_pageviews_mobile_web = {"project" : "en.wikipedia.org",
                    "access" : "mobile-web",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    "end" : '2020090100'
                        }

# Customize headers with my information
headers = {
    'User-Agent': 'https://github.com/abolim',
    'From': 'abolim@uw.edu'
}


# Function to call the APIs using the parameters created above
def api_call(endpoint,parameters):
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    
    return response
```

Let us now call the function to get the API data and perform some quality checks.

```python
# Get data in JSON format from each API call...

data_pagecounts_desktop_site = api_call(endpoint_legacy, params_pagecounts_desktop_site)
data_pagecounts_mobile_site = api_call(endpoint_legacy, params_pagecounts_mobile_site)
data_pageviews_desktop = api_call(endpoint_pageviews, params_pageviews_desktop)
data_pageviews_mobile_app = api_call(endpoint_pageviews, params_pageviews_mobile_app)
data_pageviews_mobile_web = api_call(endpoint_pageviews, params_pageviews_mobile_web)

data_pagecounts_mobile_site #QC
```

<li>Now we will save the 5 API queries as JSON files to local and read them again for further processing. </li>
<li>The JSON-formatted source data files contain the complete and un-edited output of our API queries. </li>
<li>The naming convention for the source data files is: apiname_accesstype_firstmonth-lastmonth.json </li>

```python
# Writing JSON files to local...

#JSON 1: pagecounts_desktop-site_200801-201607.json
with open('pagecounts_desktop-site_200801-201607.json', 'w') as json_file:
    json.dump(data_pagecounts_desktop_site, json_file)

#JSON 2: pagecounts_mobile-site_200801-201607.json
with open('pagecounts_mobile-site_200801-201607.json', 'w') as json_file:
    json.dump(data_pagecounts_mobile_site, json_file)

#JSON 3: pageviews_desktop_201507-202008.json
with open('pageviews_desktop_201507-202008.json', 'w') as json_file:
    json.dump(data_pageviews_desktop, json_file)

#JSON 4: pageviews_mobile-app_201507-202008.json
with open('pageviews_mobile-app_201507-202008.json', 'w') as json_file:
    json.dump(data_pageviews_mobile_app, json_file)

#JSON 5: pageviews_mobile-web_201507-202008.json
with open('pageviews_mobile-web_201507-202008.json', 'w') as json_file:
    json.dump(data_pageviews_mobile_web, json_file)
```

### Step 2: Processing the data

Now we will need to perform a series of processing steps on these data files in order to prepare them for analysis. 
Our goal is to create a single CSV-formatted data file that can be used in for analysis (step 3) with no significant additional processing. <br><br>

<li> Combine the monthly values for mobile-app and mobile-web to create a total mobile traffic count for each month. </li>
<li> All data: Separate the value of timestamp into four-digit year (YYYY) and two-digit month (MM) and discard values for day and hour (DDHH).

<li>Combine all data into a single CSV file with the following headers (formatting): <br><br>
    year                    (YYYY) <br>
    month                   (MM) <br>
    pagecount_all_views     (num_views) <br>
    pagecount_desktop_views (num_views) <br>
    pagecount_mobile_views  (num_views) <br>
    pageview_all_views      (num_views) <br>
    pageview_desktop_views  (num_views) <br>
    pageview_mobile_views   (num_views) <br>

**Let us read the JSON files which we had saved to the local and perform some quality checks.**

```python
# Reading JSON files from local...
    
with open('pagecounts_desktop-site_200801-201607.json') as file:
    pagecounts_desktop = json.load(file)
pagecounts_desktop_df = pd.DataFrame(pagecounts_desktop['items'])

with open('pagecounts_mobile-site_200801-201607.json') as file:
    pagecounts_mobile = json.load(file)
pagecounts_mobile_df = pd.DataFrame(pagecounts_mobile['items'])

with open('pageviews_desktop_201507-202008.json') as file:
    pageviews_desktop = json.load(file)
pageviews_desktop_df = pd.DataFrame(pageviews_desktop['items'])

with open('pageviews_mobile-app_201507-202008.json') as file:
    pageviews_mobileapp = json.load(file)
pageviews_mobileapp_df = pd.DataFrame(pageviews_mobileapp['items'])

with open('pageviews_mobile-web_201507-202008.json') as file:
    pageviews_mobileweb = json.load(file)
pageviews_mobileweb_df = pd.DataFrame(pageviews_mobileweb['items'])

# Checking data quality...
pagecounts_desktop_df.info()
## pagecounts_desktop_df.info()  #(103 rows x 5 cols)
## pagecounts_mobile_df.info()     #(21 rows x 5 cols)
## pageviews_desktop_df.info()     #(62 rows x 6 cols)
## pageviews_mobileapp_df.info()   #(62 rows x 6 cols)
## pageviews_mobileweb_df.info()   #(62 rows x 6 cols)
```

**Current page views API data:**
<li> Combine the monthly values for mobile-app and mobile-web to create a total mobile traffic count for each month. </li>
<li> Create year and month columns using the timestamp</li>
<li> Rename columns as per the above naming convention</li>

```python
# Mobile (web + app) page views file processing...

pageviews_mobile_df = pageviews_mobileapp_df.append(pageviews_mobileweb_df) #append the 2 dataframes
pageviews_mobile_df = pageviews_mobile_df.groupby(['timestamp']).sum().reset_index() #sum to get monthly mobile views
#pageviews_mobile_df #QC (62 rows × 2 columns)

pageviews_mobile_df['year'] = pageviews_mobile_df['timestamp'].str[0:4] #get year
pageviews_mobile_df['month'] = pageviews_mobile_df['timestamp'].str[4:6] #get month

pageviews_mobile_df = pageviews_mobile_df.rename(columns={"views": "pageview_mobile_views"}) #rename views
pageviews_mobile_df = pageviews_mobile_df.drop(columns=['timestamp']) #drop timestamp

#pageviews_mobile_df #QC (62 rows × 3 columns)
```

```python
# Desktop page views file processing...

#pageviews_desktop_df #QC (62 rows × 6 columns)
pageviews_desktop_df['year'] = pageviews_desktop_df['timestamp'].str[0:4] #get year
pageviews_desktop_df['month'] = pageviews_desktop_df['timestamp'].str[4:6] #get month

pageviews_desktop_df = pageviews_desktop_df.rename(columns={"views": "pageview_desktop_views"}) #rename views
pageviews_desktop_df = pageviews_desktop_df[['year', 'month', 'pageview_desktop_views']] #keep relevant columns

#pageviews_desktop_df #QC (62 rows × 3 columns)
```

**Legacy page counts API data:**
<li> Create year and month columns using the timestamp</li>
<li> Rename columns as per the above naming convention</li>

```python
# Desktop legacy page counts file processing...

#pagecounts_desktop_df #QC (103 rows × 3 columns)
pagecounts_desktop_df['year'] = pagecounts_desktop_df['timestamp'].str[0:4] #get year
pagecounts_desktop_df['month'] = pagecounts_desktop_df['timestamp'].str[4:6] #get month

pagecounts_desktop_df = pagecounts_desktop_df.rename(columns={"count": "pagecount_desktop_views"}) #rename views
pagecounts_desktop_df = pagecounts_desktop_df[['year', 'month', 'pagecount_desktop_views']] #keep relevant columns

#pagecounts_desktop_df #QC (103 rows × 3 columns)

# Mobile legacy page counts file processing...

#pagecounts_mobile_df #QC (21 rows x 5 columns)
pagecounts_mobile_df['year'] = pagecounts_mobile_df['timestamp'].str[0:4] #get year
pagecounts_mobile_df['month'] = pagecounts_mobile_df['timestamp'].str[4:6] #get month

pagecounts_mobile_df = pagecounts_mobile_df.rename(columns={"count": "pagecount_mobile_views"}) #rename views
pagecounts_mobile_df = pagecounts_mobile_df[['year', 'month', 'pagecount_mobile_views']] #keep relevant columns

#pagecounts_mobile_df #QC (21 rows × 3 columns)
```

**Combine all data into a single CSV file** 
<li> File with headers and formatting</li>
<li> Replace missing values with 0 </li>
<li> Create total views for both legacy and current APIs (total views is the sum of desktop and mobile views) </li>
<li> Save csv file to local desktop </li>

```python
# Joining all dataframes to create a single file

combined_df = pd.merge(pageviews_desktop_df, pageviews_mobile_df, how='outer', on=['year','month'])
combined_df = pd.merge(combined_df, pagecounts_desktop_df , how='outer', on=['year','month'])
combined_df = pd.merge(combined_df, pagecounts_mobile_df , how='outer', on=['year','month'])

# Replacing NaNs with 0s
combined_df = combined_df.fillna(0)

# Calculating total views and legacy pagecounts
combined_df['pagecount_all_views'] = combined_df['pagecount_desktop_views'] + combined_df['pagecount_mobile_views']
combined_df['pageview_all_views'] = combined_df['pageview_desktop_views'] + combined_df['pageview_mobile_views']

# Reordering columns
combined_df = combined_df[['year', 'month', 'pagecount_all_views', 'pagecount_desktop_views', 'pagecount_mobile_views', 'pageview_all_views', 'pageview_desktop_views', 'pageview_mobile_views']]
combined_df #QC (152 rows × 8 columns)

# Writing final data into a CSV file...
combined_df.to_csv('en-wikipedia_traffic_200712-202008.csv', index = False)
```

### Step 3: Analyze the data

We will visualize the dataset we have created as a time series graph. <br>
<li> We will read the saved csv file from our local desktop</li>
<li> Create a date field for chart</li>
<li> Replace 0s with nulls for better visualization</li>

```python
# Reading the CSV file back...
final_df = pd.read_csv('en-wikipedia_traffic_200712-202008.csv')

# Create date field for x-axis by using year and month
final_df['period'] = final_df['year'].astype('str') + '-' + final_df['month'].astype('str')
final_df['period'] = pd.to_datetime(final_df['period'])

# Replace 0s with NULLs for better visualization
final_df.replace(0, np.nan, inplace=True)

#Quality check
final_df.info()
```

**Let us now create the time series plot as per the following specifications:**
<li>The visualization tracks 3 traffic metrics: mobile traffic, desktop traffic, and all traffic (mobile + desktop). </li>
<li>These different sources of traffic are graphed separately whenever it is possible to do so.
<li>The scale of the y-axis (number of views) is adjusted by a factor of 1,000,000
<li>The graph contains appropriate title, legend, axes labels, gridlines and caption are added to provide the context
<li> Save a .png formatted image of your final graph is generated

```python
# Multiple line plot

plt.figure(figsize=(21,7))
x = final_df['period']

# Scaling the metrics on y-axis by a factor of 1,000,000 for better readability
scale_factor = 1000000

# Writing final data into a CSV file (for quality check)
final_df.to_csv('final_df.csv', index=False)

# Dotted lines for legacy page counts metrics
plt.plot(x, final_df['pagecount_desktop_views']/scale_factor, color='green', linestyle='dashed')
plt.plot(x, final_df['pagecount_mobile_views']/scale_factor, color='blue', linestyle='dashed')
plt.plot(x, final_df['pagecount_all_views']/scale_factor, color='black', linestyle='dashed')

# Solid lines for current page views metrics
plt.plot(x, final_df['pageview_desktop_views']/scale_factor, color='green', linewidth=1, linestyle='solid')
plt.plot(x, final_df['pageview_mobile_views']/scale_factor, color='blue', linewidth=1, linestyle='solid')
plt.plot(x, final_df['pageview_all_views']/scale_factor, color='black', linewidth=1, linestyle='solid')

# Formatting the chart area
plt.grid(color='lightgrey', linestyle='--', linewidth=0.5)
plt.title('Page Views on English Wikipedia (x 1,000,000)', fontsize=18)
plt.xlabel('Period', fontsize=12, color = 'grey')
plt.ylabel('Number of Views', fontsize=12, color = 'grey')
plt.ylim([0,12000])
txt = "May 2015: A new page view took effect, which eliminated all crawler traffic. Solid lines mark new definition."
plt.figtext(0.5, 0, txt, wrap=True, horizontalalignment='center', fontsize=14, color = 'maroon')
plt.legend(['Desktop','Mobile', 'Total'], loc="upper left", fontsize=14)

# Saving the image as a .png file to local...
plt.savefig('PageViews_TimeSeries_Plot_Aboli', pad_inches = 1)
```
